## 3. Nettoyage et gestion des valeurs manquantes


```python
missing_values = df.isnull().sum()
missing_values[missing_values > 0].sort_values(ascending=False)
```

On va remplir les valeurs manquantes avec le mode (valeur la plus fréquente) pour les variables catégorielles, ou simplement la catégorie 'Unknown'.


```python
def fill_missing(df):
    df_clean = df.copy()
    for col in df_clean.columns:
        if df_clean[col].isnull().sum() > 0:
            # Remplir par 'Unknown'
            df_clean[col].fillna('Unknown', inplace=True)
    return df_clean

df_clean = fill_missing(df)
print('Valeurs manquantes après nettoyage :', df_clean.isnull().sum().sum())
```

Sauvegarde des données nettoyées pour la modélisation.


```python
df_clean.to_csv('../../data/cleaned_data.csv', index=False)
print('Données nettoyées sauvegardées.')
```
